- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
def cabin_value_handle(X_train):
    X_train['Cabin'] = X_train['Cabin'].str.split('/').str[0]
    return X_train


X_train=cabin_value_handle(X_train)
```

```python
X_train['Cabin'].value_counts()
```

```python
def drop_pass_id(X_train):
    X_train = X_train.drop(columns=["PassengerId","Name"])
    return X_train

X_train = drop_pass_id(X_train)
```

```python
# CryoSleep and VIP has bool value but datatype is object so converting it to bool
def bool_type_conversion(X_train):
    X_train["CryoSleep"] = X_train["CryoSleep"].astype(bool)
    X_train["VIP"] = X_train["VIP"].astype(bool)
    return X_train

X_train = bool_type_conversion(X_train)
```
